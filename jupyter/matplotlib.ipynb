```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y , 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title ('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes ([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes.set_title('SMALLER PLOT')

axes2.plot(y,x)
axes2.set_title ('Larger Plot')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x **2, label = 'X Squared' )
ax.plot (x, x ** 3, label = 'X Cubed')

ax.legend( loc = (0.1,0.1))
```

```python
fig.savefig('my_picutre.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot (x,y, color = 'Purple' , lw = 2, ls = '--')

ax.set_xlim([0,1])

ax.set_ylim([0,2])
```
